```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Mean Temperature in Delhi Over the Years

```python
sns.lineplot(x=df['date'],y=df['meantemp'])
plt.show()
```

Now lets have a look at the humidity in Delhi over the years:

```python
sns.lineplot(x=df['date'],y=df['humidity'])
plt.show()
```

Wind Speed in Delhi Over the Years

```python
sns.lineplot(x=df['date'],y=df['wind_speed'])
plt.plot()
```

Now lets have a look at the relationship between temperature and humidity:

```python
plt.scatter(x=df['humidity'],y=df['meantemp'])
```

Theres a negative correlation between temperature and humidity in Delhi. It means higher temperature results in low humidity and lower temperature results in high humidity.
