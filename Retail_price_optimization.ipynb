Machine Learning models


```python
y = df['total_price']
x = df.drop('total_price',axis=1)
print(x.shape,y.shape)
```

Train_test_split

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.33,random_state=2)
```

```python
print(xtrain.shape,ytrain.shape,xtest.shape,ytest.shape)
```

Scaling

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
xtrain = sc_x.fit_transform(xtrain)
xtest = sc_x.transform(xtest)
```

decision tree regression


```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(xtrain,ytrain)
```

```python
ypred = dtr.predict(xtest)
print(ypred)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(ypred,ytest)
```

```python
plt.title='Predicted vs. Actual Retail Price',
plt.xlabel='Actual Retail Price',
plt.ylabel='Predicted Retail Price'
plt.scatter(x=ytest, y=ypred,color='#98cc89',marker='o')

x=[min(ytest), max(ytest)]
y=[min(ytest), max(ytest)]
plt.plot(x,y,marker='*')

plt.show()
```
